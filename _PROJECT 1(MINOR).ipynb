```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv('/content/drive/MyDrive/train.csv')
df
```

```python
df.info()
```

```python
df.duplicated().sum()
```

```python
df.isnull().sum()
```

```python
df["Survived"] = df["Survived"].astype("str")
df["Survived"] = df["Survived"].str.replace("1", "Survived")
df["Survived"] = df["Survived"].str.replace("0", "Died")
df = df.rename(columns = {"Survived":"Survival Status"})
df
```

```python
sns.set_style("whitegrid")
plt.figure(figsize=(11,7))
df["Survival Status"].value_counts().plot(kind = "bar")
plt.xticks(rotation = "horizontal", size = 12)
plt.xlabel("Survival Status", size = 14)
plt.show()
```

```python
df["Age"]
```

```python
plt.figure(figsize=(10,7))
sns.distplot(x = df["Age"])
plt.xlabel("Age", size = 14)
plt.ylabel("Density", size = 14)
plt.xticks(size = 12)
plt.yticks(size = 12)
plt.show()
```

```python
sns.set_style("whitegrid")
plt.figure(figsize=(9,7))
sns.barplot(x=df["Survival Status"], y = df["Age"])
plt.xlabel("Survival Status", size = 14)
plt.ylabel("Age", size = 14)
plt.xticks(size = 12)
plt.yticks(size = 12)
plt.show()
```

```python
df["Sex"].value_counts(0)
```

```python
sns.set_style("whitegrid")
plt.figure(figsize=(10,7))
df["Sex"].value_counts().plot(kind = "bar")
plt.xlabel("Sex", size= 14)
plt.xticks(rotation = "horizontal", size = 12)
plt.yticks(size = 12)
plt.show()
```

```python
plt.figure(figsize = (10,7))
sns.set_style("whitegrid")
sns.countplot(x=df["Survival Status"], hue =df["Sex"])
plt.xlabel("Survival Status", size= 14)
plt.ylabel("Count", size= 14)
plt.xticks(size = 12)
plt.yticks(size = 12)
plt.show()
```

```python
plt.figure(figsize=(16,7))
sns.distplot(x = df["Fare"])
plt.xticks(np.arange(0,600,20), size = 12)
plt.yticks(size = 12)
plt.xlabel("Fare", size= 14)
plt.ylabel("Density", size = 14)
plt.show()
```

```python
sns.set_style("whitegrid")
plt.figure(figsize=(9,7))
sns.scatterplot(x=df["Fare"], y= df["Survival Status"])
plt.xlabel("Fare", size= 14)
plt.ylabel("Survival Status", size= 14)
plt.xticks(size = 12)
plt.yticks(size = 12)
plt.show()
```
